## Open DB

```python
import pandas as pd
import sqlite3
import ast
import datetime as datetime

# Connect to your database
conn = sqlite3.connect("../tweets.db")  # adjust path if needed

# Load a table (e.g., 'tweets') into a DataFrame
df = pd.read_sql_query("SELECT * FROM tweets", conn)

# correct formatting
df['created_at'] = pd.to_datetime(df['created_at'], utc=True)

# Close connection
conn.close()
```

## Add maximum stats

```python
# add most recent views / replies etc
fields = ['likes_series', 'retweets_series', 'replies_series', 'views_series', 'engagement_timestamps']
for field in fields:
    # for val in df[field]:
    val_list = list(df[field])
    if [] in val_list:
        val_list.remove([])
    field_max = [ast.literal_eval(f)[-1]for f in val_list]
    df['max_' + field[:field.find('_')]] = field_max
```

## View most retweeted in past X hours

```python
num_tweets = 10
hours = 24

recent_tweets = df[pd.to_datetime(df['created_at']) > (datetime.datetime.now(tz=datetime.timezone.utc) - datetime.timedelta(hours=hours))]

for _, row in recent_tweets.sort_values('max_retweets', ascending=False).iloc[:num_tweets].iterrows():
    # print(row)
    print(row['created_at'])
    print(row['text'])
```

## Filter by topics

```python
topics = {
    'china': ['china', 'tariff', 'china:'],
    'inflation': ['inflation', 'cpi'],
    'trump': ['trump', 'trump:']
}

for topic in topics.keys():
    print('\n')
    print(topic)
    for idx, row in recent_tweets.iterrows():
        if len(set(topics[topic]).intersection(set(row['text'].lower().split()))) > 0:
            print(row['text'])
    
```

## Word cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all sentences into a single string
text = ' '.join(df['text'])

# Create the WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
